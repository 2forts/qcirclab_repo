```python
# Uncomment in Colab
# !pip install -q git+https://github.com/YOUR_USER/qcirclab.git

from qcirclab import Circuit, bell_pair, qft, cuccaro_adder, basic_metrics
```

```python
qc = bell_pair()
print(qc.draw())
result = qc.run(shots=512, seed=1)
print(result.counts)
print(basic_metrics(qc))
```

```python
qc = Circuit(1)
qc.h(0)
print(qc.statevector())
```

```python
qc = Circuit(2, 1)
qc.h(0)
qc.measure(0, 0)
qc.x(1, condition=(0, 1))
print(qc.draw())
print(qc.run(shots=20, seed=4).memory)
```

```python
adder = cuccaro_adder(2)
qc = Circuit(adder.n_qubits)
qc.initialize_basis('01100')
qc.append(adder)
print(qc.draw())
print(qc.statevector())
```

```python
qc = Circuit(3)
qc.initialize_basis('101')
qc.append(qft(3))
print(qc.draw())
print(qc.statevector())
```
